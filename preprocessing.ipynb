***Очистка данных от пропусков и переименование колонок на английский язык согласно требованиям***

```python
import pandas as pd
```

```python
df = pd.read_csv('Data/_data.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head(3)
```

```python
# Создадим словарь для переименования колонок на английском языке с нижним подчёркиванием

rename_columns = {
    'Unnamed: 0': 'Index',
    'ID объявления': 'ID_Ad',
    'Количество комнат': 'Rooms',
    'Тип': 'Type',
    'Метро': 'Metro',
    'Адрес': 'Address',
    'Площадь, м2': 'Area_m2',
    'Дом': 'Building',
    'Парковка': 'Parking',
    'Цена': 'Price',
    'Телефоны': 'Phones',
    'Описание': 'Description',
    'Ремонт': 'Renovation',
    'Площадь комнат, м2': 'Room_area_m2',
    'Балкон': 'Balcony',
    'Окна': 'Windows',
    'Санузел': 'Bathroom',
    'Можно с детьми/животными': 'Children_pets_allowed',
    'Дополнительно': 'Additional_info',
    'Название ЖК': 'Residential_complex',
    'Серия дома': 'Building_series',
    'Высота потолков, м': 'Ceiling_height_m',
    'Лифт': 'Elevator',
    'Мусоропровод': 'Trash_chute',
    'Ссылка на объявление': 'Listing_url'
}
```

```python
# Поменяли названия столбцов в соответствии со значениями словаря

df_rename = df.rename(columns=rename_columns)
```

```python
# Создали новый файл для удобства

df_rename.to_csv('new_data.csv', index=False)
```

```python
df_new = pd.read_csv('Data/new_data.csv')
```

```python
# Убедились, что переименование столбцов прошло успешно

df_new.head(1)
```

```python
# Посмотрим количество "плохих" данных

df_new.isna().sum()
```

```python
# А если в % ?!

(df_new.isna().sum() / len(df_new)) * 100
```

```python
# Заменяем NaN, None и пустые строки на "Нет данных"

df_new.replace(to_replace=[None, ''], value='Нет данных', inplace=True)
df_new.fillna('Нет данных', inplace=True)
```
